```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import colors as mcolors
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merge_df = city_df.merge(ride_df, left_on = 'city', right_on = 'city',sort=True)
# Display the data table for preview
merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
#Generate X and Y and Size  for "Urban"

urban_df =  merge_df[(merge_df['type']== "Urban")]

urban_df_y= urban_df.groupby('city').mean()['fare']
urban_df_y = urban_df_y.to_frame().reset_index()
urban_df_y = urban_df_y.rename(columns={"fare": "Urban"})
urban_df_y = urban_df_y.round(2)

urban_df_x = urban_df['city'].value_counts()
urban_df_x = urban_df_x.to_frame().reset_index()
urban_df_x = urban_df_x.rename(index=str, columns={"index": "city", "city": "rides"})


urban_df_size = urban_df.groupby('city').mean()['driver_count']
urban_df_size = urban_df_size.to_frame().reset_index()

```

```python
#Generate X & Y and Size for "Suburban"

Suburban_df =  merge_df[(merge_df['type']== "Suburban")]

#urban_df_x= merge_df.sort_values("fare")
Suburban_df_y= Suburban_df.groupby('city').mean()['fare']
Suburban_df_y = Suburban_df_y.to_frame().reset_index()
Suburban_df_y = Suburban_df_y.rename(columns={"fare": "Suburban"})
Suburban_df_y = Suburban_df_y.round(2)

Suburban_df_x = Suburban_df['city'].value_counts()
Suburban_df_x = Suburban_df_x.to_frame().reset_index()
Suburban_df_x = Suburban_df_x.rename(index=str, columns={"index": "city", "city": "rides"})
Suburban_df_x

Suburban_df_size = Suburban_df.groupby('city').mean()['driver_count']
Suburban_df_size = Suburban_df_size.to_frame().reset_index()
```

```python
#Generate X & Y and Size for "Rural"

rural_df =  merge_df[(merge_df['type']== "Rural")]

rural_df_y= rural_df.groupby('city').mean()['fare']
rural_df_y = rural_df_y.to_frame().reset_index()
rural_df_y = rural_df_y.rename(columns={"fare": "Rural"})
rural_df_y = rural_df_y.round(2)

rural_df_x = rural_df['city'].value_counts()
rural_df_x = rural_df_x.to_frame().reset_index()
rural_df_x = rural_df_x.rename(index=str, columns={"index": "city", "city": "rides"})

rural_df_size = rural_df.groupby('city').mean()['driver_count']
rural_df_size = rural_df_size.to_frame().reset_index()
```

```python
# Show plot
plt.scatter(urban_df_x['rides'], urban_df_y['Urban'], marker="o", facecolors="lightcoral", edgecolors="black",\
            s = urban_df_size['driver_count'] *7, alpha = 0.75)
plt.scatter(Suburban_df_x['rides'], Suburban_df_y['Suburban'], marker="o", facecolors="gold", edgecolors="blue",\
           s= Suburban_df_size['driver_count']*7 ,alpha = 0.75)
plt.scatter(rural_df_x['rides'], rural_df_y['Rural'], marker="o", facecolors="lightskyblue", edgecolors="red",\
            s= rural_df_size['driver_count']*7 ,alpha = 1)
plt.grid(True)
plt.legend(title = "City Types")
plt.title("Pyber Ride Sharing Data(2018)")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare ($)")
titletext = 'Note:\nCircle size corelates to driver count per city'
plt.suptitle(titletext, x=1, y=.8, horizontalalignment='left', verticalalignment='top', fontsize = 10)
plt.tight_layout()
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
rural_df =  merge_df[(merge_df['type']== "Rural")]
rural_df_fare= rural_df.groupby('city').sum()['fare']
rural_df_fare.sum()

Suburban_df =  merge_df[(merge_df['type']== "Suburban")]
Suburban_df_fare= Suburban_df.groupby('city').sum()['fare']
Suburban_df_fare.sum()

urban_df =  merge_df[(merge_df['type']== "Urban")]
urban_df_fare= urban_df.groupby('city').sum()['fare']
urban_df_fare.sum()


# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Urban", "Rural", "Suburban",]

# The values of each section of the pie chart
sizes = [urban_df_fare.sum(), rural_df_fare.sum(),Suburban_df_fare.sum()]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)

fig1,ax1 = plt.subplots()
ax1.set_aspect(0.75)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.title("Total Fares by City Type")
plt.tight_layout()
plt.show()
# Save Figure
plt.savefig('FaresbyCity.png')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
# Calculate Type Percents
rural_df =  merge_df[(merge_df['type']== "Rural")]
rural_df_rides= rural_df.groupby('city').count()['ride_id']


Suburban_df =  merge_df[(merge_df['type']== "Suburban")]
Suburban_df_rides= Suburban_df.groupby('city').count()['ride_id']


urban_df =  merge_df[(merge_df['type']== "Urban")]
urban_df_rides= urban_df.groupby('city').count()['ride_id']


# Build Pie Chart
labels = ["Urban", "Rural", "Suburban",]

# The values of each section of the pie chart
sizes = [urban_df_rides.sum(), rural_df_rides.sum(),Suburban_df_rides.sum()]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)
# fig1.title("")
fig1,ax1 = plt.subplots()
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
#ply.pie(title = "new title")
ax1.set_aspect(0.8)

plt.title("Total Rides by City Type")
plt.tight_layout()
plt.show()
# Save Figure
plt.savefig('RidesbyCity.png')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
rural_df =  merge_df[(merge_df['type']== "Rural")]
rural_df_drivers = rural_df.groupby('city').mean()['driver_count']

suburban_df =  merge_df[(merge_df['type']== "Suburban")]
suburban_df_drivers = suburban_df.groupby('city').mean()['driver_count']

urban_df =  merge_df[(merge_df['type']== "Urban")]
urban_df_drivers = urban_df.groupby('city').mean()['driver_count']


# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]

# The values of each section of the pie chart
sizes = [urban_df_drivers.sum(), rural_df_drivers.sum(),suburban_df_drivers.sum()]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0, 0)
# fig1.title("")
fig1,ax1 = plt.subplots()
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
#ply.pie(title = "new title")
ax1.set_aspect(0.6)

plt.title("Total Drivers by City Type")
plt.tight_layout()
plt.show()
# Save Figure
plt.savefig('DriversbyCity.png')

# Save Figure
```
